```python
import subprocess
import platform
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
# from dotenv import load_dotenv
import csv
import pandas as pd


def find_ip(ip_to_find):
    # Read the CSV file
    with open("new.csv", mode="r") as file:
        reader = csv.reader(file)
        for row in reader:
            if row[0] == ip_to_find:
                return True
        else:
            return False
        
def write_data(ip_address):
    # Define the CSV file path
    is_ip = find_ip(ip_address)
    if is_ip:
        return "Already IP exist"
    else:

        csv_file = "new.csv"
        
        # Write the IP address to the CSV file
        with open(csv_file, mode="a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([ip_address])

        # Print a success message
        print(f"IP address '{ip_address}' saved to {csv_file}")

def delete_ip(host):
    is_ip = find_ip(host)
    if is_ip:
        import pandas as pd
        # Read the CSV file without header
        df = pd.read_csv("new.csv", header=None)
        # Delete rows with value "10.3.10.50"
        df = df[df[0] != host]
        # Save the modified DataFrame back to the CSV file
        df.to_csv("new.csv", index=False, header=False)
        print(f"IP address {host} deleted from the CSV file.")
    else:
        print("IP address not found")

# Name of the cookie file for WhatsApp
whatsapp_file_cookie_name = ""

# Path to the Chrome profile
chrome_profile_path = "user-data-dir=C:/Users/Administrator/AppData/Local/Google/Chrome/User Data/what"

# Function to ping a host
def ping(host):
    """
    Returns True if the host responds to a ping request.
    """
    param = '-n' if platform.system().lower() == 'windows' else '-c'
    command = ['ping', param, '1', host]
    return subprocess.call(command) == 0

# List of hostnames to ping
hostname = ["10.10.1.117","10.100.1.5","10.100.1.1","10.100.1.20"]

# Main loop to continuously ping hosts
while True:
    for host in hostname:
        if ping(host):
            print(f"{host} is online")
            ifServer = find_ip(host)
            print(ifServer)
            if ifServer:
                delete_ip(host)
                        
        else:
            print(f"{host} is offline")
            ip = ""
            ifServer = find_ip(host)
            print(ifServer)
            print("already exist")
            # for ip in ifServer:
            #     db_ip = ip
            #     if db_ip == host:
            #         print("already exist")        
            if ifServer:
                print("Already exist in db")                
            else:
                group_id = "BaRxMtgOLVx5vDPIOMNjQt"
                # time.sleep(5)
                whatsapp_file_cookie_name = ""

                chrome_profile_path = "user-data-dir=C:/Users/Administrator/AppData/Local/Google/Chrome/User Data/what"

                chrome_options = webdriver.ChromeOptions()
                chrome_options.add_argument(chrome_profile_path)
                driver = webdriver.Chrome(options=chrome_options)
                driver.get(f"https://web.whatsapp.com/accept?code={group_id}")
                print("hello from else")
                time.sleep(120)
                # Clear field to empty it from any previous data
                driver.find_element(By.XPATH, "/html/body/div[1]/div/div[2]/div[4]/div/footer/div[1]/div/span[2]/div/div[2]/div[1]/div[2]/div[1]").clear()

                # Enter Text
                driver.find_element(By.XPATH, "/html/body/div[1]/div/div[2]/div[4]/div/footer/div[1]/div/span[2]/div/div[2]/div[1]/div[2]/div[1]").send_keys(f"This IP address {host} is not responding to pings" )
                # Use find_element instead of find_elements
                element = driver.find_element(by=By.XPATH, value="/html/body/div[1]/div/div[2]/div[4]/div/footer/div[1]/div/span[2]/div/div[2]/div[2]/button")
                element.click()  # Now you can call .click() on the single element
                time.sleep(40)
                write_data(host)
                driver.quit()
    time.sleep(10)
```
